```python
# OTIMIZATION PROBLEMS

import scipy as sp

# OR

! pip install --user scipy

# This library focus on minimization problems
# For maximization problems we need to define the otimization prices as negative

# Uma fábrica de mobilia tem à sua disposição: 75 kg de madeira de castanheiro e 100 kg de madeira de ébano
# Cada cama é vendida por 500€ e utiliza 1kg de castanheiro e 2kg de ébano
# Cada cómoda é vendida por 400€ e utiliza 1,5kg de castanheiro e 1kg de ébano
# Como é que a fábrica deve usar as matérias primas de forma a obter o máximo de receita possivel?

c = [-500, -400]
A = [[1, 1.5], [2, 1]]
b = [75, 100]

res = sp.optimize.linprog(c, A_ub=A, b_ub=b, method='revised simplex')
print(res)
```
