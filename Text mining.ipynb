```python
from bs4 import BeautifulSoup
import requests
```

```python
# Lista de músicos de jazz

jazz_musicians = ['Charlie_Parker',
                  'Dizzy_Gillespie',
                  'Art_Tatum',
                  'Clark_Terry',
                  'Dave_Brubeck',
                  'Thelonious_Monk',
                  'Charles_Mingus',
                  'Benny_Goodman',
                  'Count_Basie',
                  'John_Coltrane',
                  'Miles_Davis',
                  'Sun_Ra',
                  'Nina_Simone',
                  'Fats_Waller',
                  'Duke_Ellington',
                  'Louis_Armstrong']


# Extração da informação web-scraping  atráves do uso da biblioteca Beautiful Soup - extrair dados de um html ou xml files

documents = {}

for musician in jazz_musicians:
    
    r = requests.get('https://en.wikipedia.org/wiki/' + musician)
    soup = BeautifulSoup(r.content)
    
    main_div = soup.find('div', attrs={'class':'mw-parser-output'})
    paragraphs = main_div.find_all('p')
    paragraphs = [p.get_text() for p in paragraphs]
    documents[musician] = ''.join(paragraphs)
    
print(documents)
```

```python
# Ver um exemplo de texto
# Verificar wiki - https://en.wikipedia.org/wiki/Charlie_Parker

parker = documents['Charlie_Parker'][0:500]
print(parker)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Importação de 2 nltk packages
# Punkt = Sentence Tokenizer - divide um texto numa lista de frases, usando um algoritmo não supervisionado 
# para construir um modelo para abreviações de palavras, colocações e palavras que iniciam frases
# stopwords = dicionário de palavras
nltk.download(['punkt', 'stopwords'])
```

```python
# Tokanização do texto

parker_tokenized = word_tokenize(parker)
print(parker_tokenized)
```

```python
# Eliminar stopwords

eng_stopwords = set(stopwords.words('english'))
parker_wo_stopwords = [t for t in parker_tokenized if t not in eng_stopwords]
print(parker_wo_stopwords)
```

```python
# Eliminar pontuação

punctuation = set('&%$()"<>.,;:?!-[]{}–')
parker_wo_stopwords_and_punct = [t for t in parker_wo_stopwords if t not in punctuation]
print(parker_wo_stopwords_and_punct)
```

```python
from nltk.stem import SnowballStemmer
```

```python
# Aplicação de stemming - é o processo de reduzir a inflexão das palavras às suas formas de raiz,

stemmer = SnowballStemmer('english')
parker_wo_stopwords_and_punct_stemmed = [stemmer.stem(t) for t in parker_wo_stopwords_and_punct]
print(parker_wo_stopwords_and_punct_stemmed)
```

```python
# Aplicação das técnicas pre-processamento em todo o documento
processed_corpus = {}

# Stopwords
eng_stopwords = set(stopwords.words('english'))

# Eliminação da caracteres
punctuation = set('&%$()"<>.,;:?!-[]{}–')

# Steaming
stemmer = SnowballStemmer('english')

def process_raw_text(d):
    result = word_tokenize(d)
    result = [t for t in result if t not in eng_stopwords]
    result = [t for t in result if t not in punctuation]
    result = [stemmer.stem(t) for t in result]
    return result

for k, v in documents.items():
    processed_corpus[k] = process_raw_text(v)
```

```python
# Função para contagem de palavras únicas

unique_words = set()

for d in processed_corpus.values():
    unique_words = unique_words.union(d)

len(unique_words)
```

```python
# Definição dos parâmetros da métrica TDIDF - frequency–inverse document frequency
# Métrica usada para medir a importância de uma palavra num documento

import math

def tf(t, d):
    return d.count(t)

def idf(t, D):
    docs_with_t = sum([1 for d in D.values() if t in d])
    return math.log(len(D) / docs_with_t)

def tfidf(t, d, D):
    return tf(t, d) * idf(t, D)
```

```python
# Aplicação da função no texto para obter o valor de importância de cada palavra

tfidf_dict = {}
D = processed_corpus

for k, d in processed_corpus.items():
    
    tfidf_dict[k] = {}
    
    for t in unique_words:
        tfidf_dict[k][t] = tfidf(t, d, D)
```

```python
# Ver tabela com valor de importância para cada palavra e respectivo músico

import pandas as pd

df = pd.DataFrame(tfidf_dict)
df_t = df.transpose()
df_t
```

```python
# Ver as palavras do músico Duke_Ellington

df.Duke_Ellington.sort_values(ascending=False).head(20)
```

```python
import numpy as np

# Calcular simularidade (entre 0 e 1) entre músicos

def cosine_similarity(a, b):
    
    dot_product = np.linalg.multi_dot([a, b])
    norm_product = np.linalg.norm(a) * np.linalg.norm(a)
  
    return dot_product / norm_product
```

Quem é o músico mais parecido com o Duke Ellington?

```python
df.apply(lambda x: cosine_similarity(x, df.Duke_Ellington)).sort_values(ascending=False)
```

A similaridade pode ser usada para criar um motor de pesquisa básico. Podemos processar uma query-"Famous jazz saxophonist born in Kansas who played bebop" por exemplo-da mesma forma que processamos os dados crus e depois usar a cosine similarity para encontrar o melhor match.

```python
# Construção de um pequeno motor de pesquisa

query = 'The most influential trumpeter in jazz history'
query_after_preproc = process_raw_text(query)
query_vector = pd.Series({t:tfidf(t, query_after_preproc, D) for t in unique_words})
```

```python
df.apply(lambda x: cosine_similarity(x, query_vector)).sort_values(ascending=False)
```
